```
f = open("data.txt", 'r')
```

```
samp = f.read()
samp = samp.split("\n")
```

```
samp
```

```
size, time = [], []
for i in samp:
  size.append(int(i.split(',')[0]))
  time.append(float(i.split(',')[1]))
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(size, time, test_size = 0.2, random_state = 101)
```

```
import matplotlib.pyplot as plt
plt.scatter(size, time)
```

```
import pandas as p
x_train = p.DataFrame(x_train)
x_test = p.DataFrame(x_test)
```

```
x_train
```

```
x_test
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```
predict = model.predict(x_test)
```

```
model.score(x_test, y_test)
```

```
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, predict)
```
